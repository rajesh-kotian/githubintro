# Use Pillow to open Image

```python
from PIL import Image
alic_mask = np.array(Image.open("Alice_mask.png"))
alic_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
len(alic_mask)
```

```python
transformed_alic_mask = np.ndarray((alic_mask.shape[0], alic_mask.shape[1]), np.int32)

alic_mask.shape[0]
alic_mask.shape[1]
```

```python
transformed_alic_mask = np.ndarray((alic_mask.shape[0], alic_mask.shape[1]), np.int32)

for i in range(len(alic_mask)):
    transformed_alic_mask[i] = list(map(transform_format,alic_mask[i]))
    
transformed_alic_mask
```

```python
# Create a word cloud image
wc  = WordCloud(background_color="white" , max_words = 1000,
               mask = transformed_alic_mask,
               stopwords=stopwords, contour_width = 3,
               contour_color = 'blue')
# Generate a wordcloud
wc.generate(text1)

#store to file
wc.to_file("alice.png")

# show 
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
